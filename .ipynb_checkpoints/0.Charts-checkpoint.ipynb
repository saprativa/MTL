```python
import pickle
with open('facebook-bart-base.pickle', 'rb') as f:
    training_results1 = pickle.load(f)
    
with open('facebook-bart-large-cnn.pickle', 'rb') as f:
    training_results2 = pickle.load(f)
    
with open('facebook-bart-large-xsum.pickle', 'rb') as f:
    training_results3 = pickle.load(f)
    
with open('philschmid-bart-base-samsum.pickle', 'rb') as f:
    training_results4 = pickle.load(f)
    
with open('Salesforce-bart-large-xsum-samsum.pickle', 'rb') as f:
    training_results5 = pickle.load(f)
    
with open('sshleifer-distilbart-cnn-6-6.pickle', 'rb') as f:
    training_results6 = pickle.load(f)
    
with open('allenai-led-base-16384.pickle', 'rb') as f:
    training_results7 = pickle.load(f)
```

```python
import matplotlib.pylab as plt
plt.plot(training_results1[0]['training_loss'], label="facebook-bart-base")
plt.plot(training_results2[0]['training_loss'], label="facebook-bart-large-cnn")
plt.plot(training_results3[0]['training_loss'], label="facebook-bart-large-xsum")
plt.plot(training_results4[0]['training_loss'], label="philschmid-bart-base-samsum")
plt.plot(training_results5[0]['training_loss'], label="Salesforce-bart-large-xsum-samsum")
plt.plot(training_results6[0]['training_loss'], label="sshleifer-distilbart-cnn-6-6")
plt.title('Training loss')
plt.ylabel('loss')
plt.xlabel('iterations')
plt.legend()
```

```python
plt.plot(training_results1[0]['validation_loss'], label="facebook-bart-base")
plt.plot(training_results2[0]['validation_loss'], label="facebook-bart-large-cnn")
plt.plot(training_results3[0]['validation_loss'], label="facebook-bart-large-xsum")
plt.plot(training_results4[0]['validation_loss'], label="philschmid-bart-base-samsum")
plt.plot(training_results5[0]['validation_loss'], label="Salesforce-bart-large-xsum-samsum")
plt.plot(training_results6[0]['validation_loss'], label="sshleifer-distilbart-cnn-6-6")
plt.title('Validation loss')
plt.ylabel('loss')
plt.xlabel('iterations')
plt.legend()
```

```python
import matplotlib.pylab as plt
plt.plot(training_results6[0]['training_loss'], label="sshleifer-distilbart-cnn-6-6")
plt.plot(training_results7[0]['training_loss'], label="allenai-led-base-16384")
plt.title('Training loss')
plt.ylabel('loss')
plt.xlabel('iterations')
plt.legend()
```
